# MEANING OF EACH PARAMETERS
**Date**: The date of the data entry.  
**Country/Region**: The country or region name.  
**Confirmed**: Total confirmed cases.  
**Deaths**: Total deaths.  
**Recovered**: Total recovered cases.  
**Active**: Active cases (Confirmed - Deaths - Recovered).  
**New cases**: New confirmed cases on the given date.  
**New deaths**: New deaths on the given date.  
**New recovered**: New recoveries on the given date.  
**WHO Region**: WHO region classification.
