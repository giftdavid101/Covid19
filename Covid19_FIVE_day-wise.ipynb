# MEANING OF COLUMN PARAMETERS

**Date**: Records the specific date.  
**Confirmed**: Shows cumulative confirmed cases.  
**Deaths**: Shows cumulative deaths.  
**Recovered**: Shows cumulative recoveries.  
**Active**: Represents current active cases.  
**New cases**: Tracks new confirmed cases for the date.  
**New deaths**: Tracks new deaths for the date.  
**New recovered**: Tracks new recoveries for the date.  
**Deaths / 100 Cases**: Indicates death rate per 100 confirmed cases.  
**Recovered / 100 Cases**: Indicates recovery rate per 100 confirmed cases.  
**Deaths / 100 Recovered**: Indicates death rate per 100 recovered cases.  
**No. of countries**: The number of reporting countries.
